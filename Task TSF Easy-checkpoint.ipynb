```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt  
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```


<font color='BurlyWood'>
<h2>First We Read the file from the URL Provided</h2>
</font>


```python
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
```

```python
data
```


<font color='BurlyWood'>
<h2>Plotting Hours W.R.T Scores</h2>
</font>


```python
data.plot()
```

```python
data.plot(kind='scatter',x='Hours',y='Scores')
```

<font color='Coral'>
<h2>Pie Chart Gives a better understanding for the given DATA SET</h2>
</font>

```python
plt.figure(figsize=(13,6))
plt.pie(data.Hours, labels = data.Scores, 
        startangle=90, shadow = True,
        radius = 1.5,autopct='%1.2f%%')
plt.legend()
plt.show()
```

<font color='BurlyWood'>
<h2>Splitting Up the Data</h2>
</font>

```python
from IPython import display
display.Image("C:\\Users\\hp\\Desktop\\goku.gif")
```

```python
x_train, x_test, y_train, y_test = train_test_split(data.Hours, data.Scores, test_size = 0.2)
```

<font color='BurlyWood'>
<h2>Implement the Model Now</h2>
</font>

```python
regr = LinearRegression()
```

```python
regr.fit(np.array(x_train).reshape(-1,1), y_train)
```

<font color='Indigo'>
<h2>One - dimensional columnar array for Predictions (using reshape(-1,1))</h2>
</font>

```python
preds = regr.predict(np.array(x_test).reshape(-1,1))
```

```python
y_test.head()
```

```python
preds
```

```python
residuals = preds - y_test
```

```python
plt.hist(residuals)
```

<font color='LightPink'>
<h2>Predicted Versus Actual in a DataFrame</h2>
</font>

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': preds})  
df 
```

<font color='Indigo'>
<h2>Prediction For a Specific Value using predict() function</h2>
</font>

```python
hours = [[9.25]]
own_pred = regr.predict(hours)
```

```python
print("No of Hours = {}".format(hours))
print("Predicted Score = {}".format(own_pred[0]))
```
